```python
# Variables (Data Types)

book_title = "Python Programming"
book_quantity = 5
book_price = 29.99
is_available = True

print(f"Book: {book_title}, Quantity: {book_quantity}, Price: ${book_price}, Available: {is_available}")
```

```python
# Lists

books = ["Python Basics", "Advanced Python", "Data Science with Python"]

# Add two more books
books.append("Machine Learning Essentials")
books.insert(1, "Python for Beginners")

print("Updated book list:", books)
```

```python
# Loops

print("\nAvailable books:")
for i in range(len(books)):
    print(f"{i}. {books[i]}")
```

```python
# User Input

new_book = input("Enter a new book title: ")
books.append(new_book)
print("Updated book list:", books)
```

```python
# Functions (Check Availability)

def check_availability(book_name):
    return book_name in books

print(f"Is 'Python Basics' available? {check_availability('Python Basics')}")
```

```python
# File I/O

def save_books_to_file(filename):
    with open(filename, 'w') as f:
        for book in books:
            f.write(f"{book}\n")

def read_books_from_file(filename):
    loaded_books = []
    try:
        with open(filename, 'r') as f:
            for line in f:
                loaded_books.append(line.strip())
    except FileNotFoundError:
        print("File not found.")
    return loaded_books

save_books_to_file("books.txt")
print("Books read from file:", read_books_from_file("books.txt"))
```

```python
# Pickle Save/Load

def save_books_pickle(filename):
    with open(filename, 'wb') as f:
        pickle.dump(books, f)

def load_books_pickle(filename):
    try:
        with open(filename, 'rb') as f:
            return pickle.load(f)
    except FileNotFoundError:
        print("Pickle file not found.")
        return []

save_books_pickle("books.pkl")
print("Books loaded from pickle:", load_books_pickle("books.pkl"))
```

```python
# SQLite Database

def create_books_table():
    conn = sqlite3.connect('library.db')
    cursor = conn.cursor()
    cursor.execute('''CREATE TABLE IF NOT EXISTS books
                      (id INTEGER PRIMARY KEY, title TEXT, quantity INTEGER)''')
    conn.commit()
    conn.close()

def add_book_to_db(title, quantity):
    conn = sqlite3.connect('library.db')
    cursor = conn.cursor()
    cursor.execute("INSERT INTO books (title, quantity) VALUES (?, ?)", (title, quantity))
    conn.commit()
    conn.close()
    print(f"Book '{title}' added to database.")

create_books_table()
add_book_to_db("Python for Beginners", 10)
```

```python
# Web API (GitHub)

def get_python_repos():
    url = "https://api.github.com/search/repositories"
    params = {"q": "language:python", "sort": "stars", "order": "desc"}
    response = requests.get(url, params=params)
    data = response.json()
    return [repo['name'] for repo in data['items'][:5]]

print("Top 5 Python repositories on GitHub:", get_python_repos())
```

```python
# User Interface

def show_menu():
    print("\n===== Library Menu =====")
    print("1. List all books")
    print("2. Add a book")
    print("3. Remove a book")
    print("4. Save books to file")
    print("5. Load books from file")
    print("6. Exit")
    print("========================")

def list_books():
    if len(books) == 0:
        print("No books available.")
    else:
        print("\nCurrent Books:")
        for i, book in enumerate(books):
            print(f"{i}. {book}")

def add_book():
    title = input("Enter book title to add: ")
    books.append(title)
    print(f"'{title}' added successfully.")

def remove_book():
    list_books()
    try:
        index = int(input("Enter the index of the book to remove: "))
        removed = books.pop(index)
        print(f"'{removed}' removed successfully.")
    except (ValueError, IndexError):
        print("Invalid index. Please try again.")

def save_books_menu():
    filename = input("Enter filename to save (e.g., books.txt): ")
    save_books_to_file(filename)
    print(f"Books saved to {filename}")

def load_books_menu():
    filename = input("Enter filename to load (e.g., books.txt): ")
    loaded = read_books_from_file(filename)
    if loaded:
        books.clear()
        books.extend(loaded)
        print("Books loaded successfully:")
        list_books()

# MAIN MENU LOOP
while True:
    show_menu()
    choice = input("Choose an option (1-6): ")

    if choice == "1":
        list_books()
    elif choice == "2":
        add_book()
    elif choice == "3":
        remove_book()
    elif choice == "4":
        save_books_menu()
    elif choice == "5":
        load_books_menu()
    elif choice == "6":
        print("Exiting Library System. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number from 1 to 6.")
```
