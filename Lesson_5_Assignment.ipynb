````python
# Task 1 — Load the Titanic Dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset from uploaded file
df = pd.read_csv("Titanic-Dataset.csv")

df.head()
````

````python
# Task 2 — Inspect the Data
# First 5 rows
df.head()

````

````python
# Last 5 rows
df.tail()
````

````python
# Data types
df.dtypes
````

````python
# Summary statistics
df.describe()
````

````python
# Task 3 — Clean & Preprocess the Data

# Convert Pclass to category
df['PclassCategory'] = df['Pclass'].astype('category')

# Create CabinCategory (first letter of cabin or 'Unknown')
df['CabinCategory'] = df['Cabin'].astype(str).str[0]
df['CabinCategory'] = df['CabinCategory'].replace('n', 'Unknown')

# Fill missing Age values with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Create Family feature
df['Family'] = df['SibSp'] + df['Parch']
````

````python
# Task 4 — Identify Specific Data Points

# Passengers with highest fare
highest_fare_passengers = df[df['Fare'] == df['Fare'].max()]
highest_fare_passengers
````

````python
# Passengers with largest family size
largest_family_passengers = df[df['Family'] == df['Family'].max()]
largest_family_passengers
````

````python
# Task 5 — Pandas Series & DataFrame Functionalities
# Series of names
names_series = pd.Series(df['Name'])
names_series.head()
````

````python
# DataFrame of names and ages
names_ages_df = df[['Name', 'Age']]
names_ages_df.head()
````

````python
# Passengers older than 50
older_passengers = df[df['Age'] > 50]
older_passengers.head()
````

````python
# Task 6 — Multi‑Indexing

multi_index_df = df.groupby(['Sex', 'Pclass'])['Survived'].mean()
multi_index_df
````

````python
# Task 7 — Visualizations
# Histogram of Ages

plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Histogram of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
````

````python
# Bar Plot — Survival by Class

plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()
````

````python
# Heatmap — Correlation

plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
````

````python
# Pair Plot

sns.pairplot(df[['Age', 'Fare', 'SibSp', 'Parch', 'Survived']], hue='Survived')
plt.show()
````

````python
# Task 8 — Analyze Survival Rates

# Survival by sex
survival_by_sex = df.groupby('Sex')['Survived'].mean()
survival_by_sex
````

````python
# Survival by class
survival_by_class = df.groupby('Pclass')['Survived'].mean()
survival_by_class
````

````python
# Task 9 — Age Distribution (Box Plot)

plt.figure(figsize=(12, 6))
sns.boxplot(x='Pclass', y='Age', hue='Sex', data=df)
plt.title('Age Distribution by Passenger Class and Sex')
plt.show()
````

````python
# Task 10 — Fare Distribution (Violin Plot)

plt.figure(figsize=(12, 6))
sns.violinplot(x='Pclass', y='Fare', hue='Survived', data=df, split=True)
plt.title('Fare Distribution by Passenger Class and Survival Status')
plt.show()
````

````python
# Markdown Notes for EDA Insights

# Survival by Sex
# Females had a significantly higher survival rate.

# This aligns with the “women and children first” evacuation policy.

# Survival by Class
# First‑class passengers survived the most.

# Third‑class passengers had the lowest survival rate.

# Age Distribution
# Younger passengers (children) had higher survival rates.

# Older passengers were less likely to survive.

# Fare Distribution
# Higher fares correlate with higher survival.

# Wealthier passengers had better access to lifeboats.

# Family Size
# Very large families had lower survival rates.

# Small families (1–3 members) had better outcomes.
````
