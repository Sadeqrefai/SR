```python
# ============================================================
# WINE DATASET – FULL EDA, SCALING, POLYNOMIAL FEATURES,
# LINEAR REGRESSION, LASSO REGRESSION
# ============================================================

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.metrics import mean_squared_error

# ------------------------------------------------------------
# 1. LOAD THE WINE DATASET
# ------------------------------------------------------------

wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df["target"] = wine.target

print("First 5 rows:")
display(df.head())

# ------------------------------------------------------------
# 2. TRAIN–TEST SPLIT (80/20)
# ------------------------------------------------------------

X = df.drop("target", axis=1)
y = df["target"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

# ------------------------------------------------------------
# 3. EXPLORATORY DATA ANALYSIS (EDA)
# ------------------------------------------------------------

print("\nDataset Info:")
print(df.info())

print("\nSummary Statistics:")
display(df.describe())

# Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Pairplot (choose 4 features + target)
sns.pairplot(df[["alcohol", "malic_acid", "ash", "hue", "target"]], hue="target")
plt.show()

# Histogram of target
plt.figure(figsize=(6, 4))
sns.countplot(x=df["target"])
plt.title("Distribution of Target Classes")
plt.show()

# ------------------------------------------------------------
# 4. FEATURE SCALING
# ------------------------------------------------------------

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

print("\nFirst 5 rows of scaled features:")
display(scaled_df.head())

# ------------------------------------------------------------
# 5. POLYNOMIAL FEATURES (Degree 2)
# ------------------------------------------------------------

poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X_scaled)

print("\nShape of polynomial features:", X_poly.shape)

poly_df = pd.DataFrame(X_poly)
print("\nFirst 5 rows of polynomial features:")
display(poly_df.head())

# ------------------------------------------------------------
# 6. MODEL TRAINING – LINEAR REGRESSION
# ------------------------------------------------------------

# Split polynomial features
X_train_poly, X_test_poly, y_train_poly, y_test_poly = train_test_split(
    X_poly, y, test_size=0.20, random_state=42
)

lin_reg = LinearRegression()
lin_reg.fit(X_train_poly, y_train_poly)

y_pred_lin = lin_reg.predict(X_test_poly)

rmse_lin = np.sqrt(mean_squared_error(y_test_poly, y_pred_lin))
print("\nLinear Regression RMSE:", rmse_lin)

# ------------------------------------------------------------
# 7. MODEL TRAINING – LASSO REGRESSION
# ------------------------------------------------------------

lasso = Lasso(alpha=0.01, max_iter=5000)
lasso.fit(X_train_poly, y_train_poly)

y_pred_lasso = lasso.predict(X_test_poly)

rmse_lasso = np.sqrt(mean_squared_error(y_test_poly, y_pred_lasso))
print("Lasso Regression RMSE:", rmse_lasso)

# ------------------------------------------------------------
# 8. MODEL COMPARISON
# ------------------------------------------------------------

print("\nModel Comparison:")
print("Linear Regression RMSE:", rmse_lin)
print("Lasso Regression RMSE:", rmse_lasso)

if rmse_lasso < rmse_lin:
    print("\nLasso performed better because it reduces overfitting by shrinking coefficients.")
else:
    print("\nLinear Regression performed better, likely because the dataset is small and clean.")
```
