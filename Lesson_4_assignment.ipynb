```python
# Task 9 — Iterators and Generators

# Iterator class
class EvenIterator:
    def __init__(self):
        self.num = 0
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < 5:
            result = self.num
            self.num += 2
            self.count += 1
            return result
        else:
            raise StopIteration

# Generator function
def even_generator():
    num = 0
    for _ in range(5):
        yield num
        num += 2

# Demonstration
print("Iterator:", list(EvenIterator()))
print("Generator:", list(even_generator()))
```

```python
#  Task 10 — Map, Reduce, Filter

numbers = [1, 2, 3, 4, 5]

squared = list(map(lambda x: x*x, numbers))
product = reduce(lambda x, y: x*y, numbers)
evens = list(filter(lambda x: x % 2 == 0, numbers))

print("Squared:", squared)
print("Product:", product)
print("Even numbers:", evens)
```

```python
# Task 11 — Rectangle Class

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

rect1 = Rectangle(5, 3)
rect2 = Rectangle(10, 4)

print("Rectangle 1 - Area:", rect1.area(), "Perimeter:", rect1.perimeter())
print("Rectangle 2 - Area:", rect2.area(), "Perimeter:", rect2.perimeter())
```
